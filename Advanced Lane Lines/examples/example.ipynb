```python
def corners_unwarp(img, nx, ny, mtx, dist):
   
    #Search for corners in the grayscaled image , using 9x6 chessboard
    ret, corners = cv2.findChessboardCorners(gray_img, (nx, ny), None)
    if ret == True:
        # Given src and dst points, calculate the perspective transform matrix
        M = cv2.getPerspectiveTransform(src, dst)
        # Warp the image using OpenCV warpPerspective()
        warped = cv2.warpPerspective(undist, M, img_size)

    # Return the resulting image and matrix
    return warped, M
        
    ret, matrix, dist_coeff, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)
    dst = cv2.undistort(img, matrix, dist_coeff, None, matrix)
    return dst

# applying Sobel
def abs_sobel_thresh(gray_img, orient='x', thresh_min=0, thresh_max=255):
    gray = cv2.cvtColor(gray_img, cv2.COLOR_RGB2GRAY)
    
    if orient == 'x':
        sobel = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
    else:
        sobel = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
        
    abs_sobel = np.absolute(sobel)
    scaled_sobel = np.uint8(255 * abs_sobel/np.max(abs_sobel))
    sobel_binary = np.zeros_like(scaled_sobel)
    sobel_binary[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1
    
    return sobel_binary

# compute magnitude
def mag_thresh(gray_img, sobel_kernel=3, mag_thresh=(0, 255)):
    
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize =  sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1,ksize =  sobel_kernel)
    abs_dist = np.sqrt(np.power(sobelx,2) + np.power(sobely,2))
    scaled_sobel = np.uint8(255 * abs_dist /np.max(abs_dist ))
    sobel_binary = np.zeros_like(scaled_sobel)
    sobel_binary[(scaled_sobel >= mag_thresh[0]) & (scaled_sobel <= mag_thresh[1])] = 1
    return sobel_binary

# compute direction
def dir_threshold(gray_img, sobel_kernel=3, thresh=(0, np.pi/2)):
    
    sobelx = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize =  sobel_kernel)
    sobely = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1,ksize =  sobel_kernel)
    abs_sobelx = np.sqrt(np.power(sobelx,2))
    abs_sobely = np.sqrt(np.power(sobely,2))
    direction = np.arctan2(abs_sobely, abs_sobelx)
    sobel_binary = np.zeros_like(direction)
    sobel_binary[(direction >= thresh[0]) & (direction <= thresh[1])] = 1
    return sobel_binary
    
```
